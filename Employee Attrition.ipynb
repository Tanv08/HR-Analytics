# What is the overall attrition rate for the company?

```python
Attrition_Yes = (data["Attrition"] == "Yes").value_counts()
Attrition_Yes
```

```python
Attrition_Rate_of_the_company = (237/1470)
Attrition_Rate_of_the_company
```

Attrition_Rate_of_the_company = 16.12%

```python
def attrition(by, ascending = False):
    df = data.loc[data['Attrition'] == 'Yes',["EmployeeNumber", by]].sort_values(by = by, ascending = ascending).copy()
    return df[["EmployeeNumber",by]] 
```

# Are employees with a certain job level more likely to leave the company?

```python
By_JobLevel = attrition(by = 'JobLevel')
By_JobLevel
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="JobLevel", data=By_JobLevel, order=By_JobLevel['JobLevel'].sort_values().unique())
plt.xticks(rotation='vertical')
plt.title('Attrition by JobLevel', fontsize=40)
plt.ylabel('Count', fontsize=25)
plt.xlabel('JobLevel', fontsize=25)
plt.show()
```

Yes, employees who are at their first job level showed more attrition than those at higher levels.

# Do employees who receive lower % Salary Hikes have a higher attrition rate?
